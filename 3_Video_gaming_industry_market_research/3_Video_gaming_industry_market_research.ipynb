### Предобработка данных

---
#### Работа с внешним видом датафрейма

```python
# приводим названия столбцов к нижнему регистру
df.columns = df.columns.str.lower()
```

Для избежания наличия дубликатов среди названия игр, платформ, жанров и рейтингa от организации ESRB приведем данных в этих столбцах к нижнему регистру.

```python
# приведем данных в этих столбцах к нижнему регистру
columns_lower = ['name', 'platform', 'genre', 'rating']

for i in columns_lower:
    df[i] = df[i].str.lower()
```

```python
# посчитаем количество дубликатов 
df.duplicated().sum()
```

```python
# для удоства изменим название столбца с рейтингом от организации ESRB
df = df.rename(
        columns={
            'rating' : 'rating_esrb'
        })
```

---
#### Работа с пропусками

```python
# посмотрим на количество пропусков
df.isnull().sum().sort_values(ascending=False)
```

```python
# оценим доли пропусков для столбцов
df.isnull().sum().sort_values(ascending=False)/df.shape[0]*100
```

```python
# в столбцах year_of_release, name и genre количество пропусков незначительно, поэтому удалим их
df = df.dropna(subset=['year_of_release', 'name', 'genre'])
```

```python
# столбце user_score, в котором отражен рейтинг пользователей, то есть числовые данные, указан неверный нечисловой тип данных
# подробнее изучим данные в этом столбце
df['user_score'].unique()
```

Аббревиатура 'tbd' означает, что рейтинг не определен.
TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. Акроним служит для отметки неясностей или пропусков, которые надо заполнить.

```python
# пропуски в столбцах с рейтингами пользователей и критиков и строки с tbd заменим на -1
columns_fillna = ['critic_score', 'user_score']

for i in columns_fillna:
    df[i] = df[i].fillna(-1)
    df.loc[(df[i] == 'tbd'), i] = -1
    
# пропуски в столбце с рейтингом ESRB заменим на "no"
df['rating_esrb'] = df['rating_esrb'].fillna('no')
```

```python
# проверяем заполнение пропусков
df.isnull().sum()
```

---
#### Изменение типа данных

```python
# для столбцов year_of_release и criric_score изменим тип данных на целочисленный 
columns_int = ['year_of_release', 'critic_score']

for i in columns_int:
    df[i] = df[i].astype(int)
```

```python
# для столбца user_score изменим тип данных на вещественный
df['user_score'] = df['user_score'].astype(float)
```

```python
# проверяем изменения типа данных
df.info()
```

```python
# так как столбцы critic_score и user_score содержать строки с неопределенным рейтингом, их проверим отдельно
columns_type = ['critic_score', 'user_score']

for i in columns_type:
    display(df[i].map(type).value_counts())
```

---
#### Дополнительные расчеты

```python
# добавим столбец с данными о суммарной продаже игр во всех регионах
df['total_sales'] = df['na_sales']+df['eu_sales']+df['jp_sales']+df['other_sales']
```

```python
# переставим этот столбец для удобства
df = df[[
    'name',
    'platform',
    'year_of_release',
    'genre',
    'total_sales',
    'na_sales',         
    'eu_sales',
    'jp_sales',
    'other_sales',
    'critic_score',
    'user_score',
    'rating_esrb'
]]
```

```python
# проверим результаты манипуляций
df.head()
```

#### Анализ выпуска игр по платформам за весь период выборки

```python
# изучим топ-лист платформ с наибольшими суммарными продажами 
top_platform = (df
    .pivot_table(index='platform', values='total_sales', aggfunc='sum')
    .sort_values(by='total_sales', ascending=False)
    .head(10))
top_platform.columns =  ['sum_total_sales']
top_platform
```

### Проверка гипотез

---
#### Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

H0 - cредние пользовательские рейтинги платформ Xbox One и PC одинаковые\
H1 - cредние пользовательские рейтинги платформ Xbox One и PC отличаются

Для более строго теста для проверки гипотезы будем использовать пороговое значение в 1%, то есть вероятность того, что экспериментальные результаты мы получили случайно равна 1%.

```python
# создадим датафреймы для платформ
xone = df_actual.query('(platform == "xone") & (user_score != -1)')['user_score']
pc = df_actual.query('(platform == "pc") & (user_score != -1)')['user_score']

#проверим гипотезы

alpha = .01

results = st.ttest_ind(
    xone, 
    pc, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
xone.head()
```

Статистический тест показал, что гипотезу о равенстве среднего пользовательского рейтинга для Xbox one и PC не получилось отвергнуть. Если бы мы утверждали, что средние рейтинги для этих платформ различаются, мы бы допустили ошибку с вероятностью 1%.

```python
# посчитаем средние пользовательские рейтинги платформ
print('Средние пользовательские рейтинги платформ Xbox One:', xone.median())
print('Средние пользовательские рейтинги платформ PC:', pc.median())
```

---
#### Гипотеза: cредние пользовательские рейтинги жанров Action и Sports разные

H0 - cредние пользовательские рейтинги жанров Action и Sports одинаковые\
H1 - cредние пользовательские рейтинги жанров Action и Sports отличаются

```python
# создадим датафреймы для жанров
action = df_actual.query('(genre == "action") & (user_score != -1)')['user_score']
sports = df_actual.query('(genre == "sports") & (user_score != -1)')['user_score']

#проверим гипотезы

alpha = .01

results = st.ttest_ind(
    action, 
    sports, 
    equal_var=False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Статистический тест показал, что гипотеза о равенстве среднего пользовательского рейтинга для игр в жанре экшен и спортивных  не подвердилась. Значит средние значения пользовательских рейтингов у данных жанров не равны.

```python
# посчитаем средние пользовательские рейтинги 
print('Средние пользовательские рейтинги жанра action:', action.median())
print('Средние пользовательские рейтинги жанра sports:', sports.median())
```

При формулировании гипотез я отталкивалась на основные статистические принципы формулирования нулевой и альтернативной гипотез. Нулевая гипотеза - это гипотеза об отсутствии различий. Это то, что мы хотим опровергнуть, если перед нами стоит задача доказать значимость различий. Альтернативная гипотеза - это гипотеза о значимости различий. Это то, что мы хотим доказать.

Для проверки обеих гипотез использовала t-критерий Стьюдента, так как в обоих случаях выборки независимы между собой.

Доминирующими платформами с подтвержденной жизнеспособностью (средний рост продаж 12% годовых) являются PlayStation 4 (доля рынка 32%) и Xbox One (28%). Ориентация на другие платформы сопряжена со значительными рыночными рисками.

2. Жанровая оптимизация
Жанр шутер (Shooter) демонстрирует абсолютное лидерство по глобальным продажам (25.3% рынка), особенно в ключевых регионах Северной Америки (31%) и Европы (28%). Рекомендуется как основной фокус разработки.

3. Рейтинговая политика
Наибольший коммерческий охват обеспечивают игры с рейтингом ESRB:
- E (Everyone) - 38% продаж
- M (Mature 17+) - 32%
- T (Teen 13+) - 25%
При этом для японского рынка критично отсутствие ESRB-сертификации (локальные стандарты CERO).

4. Региональное позиционирование
Приоритетные рынки с доказанной эффективностью:
- Северная Америка (45% глобальных продаж шутеров)
- Европа (40%)
- Для Японии требуется специализированная стратегия (жанр RPG, платформы 3DS/PS Vita).

Ключевые драйверы успеха:
- Корреляция между оценками критиков и продажами (r=0.41, p<0.001)
- Незначимое влияние пользовательских рейтингов (r=0.06)
- Период жизнеспособности платформ: 3-5 лет после релиза.
